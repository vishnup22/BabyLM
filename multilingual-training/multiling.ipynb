```python
tier1 = """zho
nld
deu
fra
ind
fas
ukr
bul""".split()

tier2 = """yue
est
swe
cym
pol
afr
eus
ita
spa
por
jpn
heb
srp
ara
ell""".split()

tier3 = """bug
hun
tur
ces
ace
dan
ban
hrv
mak
nso
ron
nor
isl
zul
sot
xho
kor
rus
sun
jav""".split()

tiers = {
    3: tier3,
    2: tier2,
    1: tier1,
}
batch_sizes = {
    1: 64,
    2: 16,
    3: 8,
}
epochs = {
    1: 5,
    2: 10,
    3: 10,
}
```

```python
import subprocess


def create_script(langcodes, model_name, config_type, batch_size, epochs, vocab_size=8192):
    assert config_type in ["small", "large", "fast", "gemma"]

    if isinstance(langcodes, list):
        dataset_str = " ".join([f"BabyLM-community/babylm-{langcode}" for langcode in langcodes])
    else:
        dataset_str = f"BabyLM-community/babylm-{langcodes}"
    
    return f"""python train.py \
        --dataset "{dataset_str}" \
        --config {config_type}_config.json \
        --output_dir ./{model_name}-baseline-{config_type} \
        --model_name BabyLM-community/{model_name}-baseline-{config_type} \
        --epochs {epochs} \
        --vocab_size {vocab_size} \
        --push_to_hub \
        --batch_size {batch_size}
    """
```

```python
# create_script_multiling
all_langs = tier1 + tier2 + tier3
script = create_script(all_langs, "ALL", "small", 64, 1, vocab_size=32768)
subprocess.run(script, shell=True, executable="/bin/bash")
```

```python
for lang in tier1[6:]:
    print(lang)
    script = create_script([lang, "eng"], f"eng-{lang}", "small", 64, 5, vocab_size=32768)
    subprocess.run(script, shell=True, executable="/bin/bash")
```
